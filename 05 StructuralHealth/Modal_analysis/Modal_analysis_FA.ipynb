Because rotations are gradient of displacements; Let us try to fix the BCs from the gradients of displacements

```python
###### Obtaining the mode shapes of differential of translations ############
##### defined here and used later in the notebook ##############
def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)    
    for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return norm_arr
range_to_normalize = (-1,1)

MS1_gradient = np.gradient(MS1_smooth, LOB[1]-LOB[0])
MS1_gradient_norm = normalize(MS1_gradient, range_to_normalize[0], range_to_normalize[1])

MS2_gradient = np.gradient(MS2_smooth, LOB[1]-LOB[0])
MS2_gradient_norm = normalize(MS2_gradient, range_to_normalize[0], range_to_normalize[1])

MS3_gradient = np.gradient(MS3_smooth, LOB[1]-LOB[0])
MS3_gradient_norm = normalize(MS3_gradient, range_to_normalize[0], range_to_normalize[1])

MS4_gradient = np.gradient(MS4_smooth, LOB[1]-LOB[0])
MS4_gradient_norm = normalize(MS4_gradient, range_to_normalize[0], range_to_normalize[1])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8,8))
# Plot the initial data
line1, = ax[0,0].plot(LOB, MSrot1_smooth, color='#d48500', linewidth=1, linestyle='-') 
ax[0,0].axhline(y=0, color='k', linestyle='-', linewidth=2)  
ax[0,0].set_xlim(0, 24)
ax[0,0].set_ylim([-1.25, 1.25])
ax[0,0].grid()
ax[0,0].set_ylabel('Normalized amplitude \n of rotation', family='serif', color='black', size=10)
#ax[0,0].set_xlabel('Length of the Bridge', family='serif', color='black', size=10)

line2, = ax[0,1].plot(LOB, MSrot2_smooth, color='#d48500', linewidth=1, linestyle='-', label='Mode I')    
ax[0,1].axhline(y=0, color='k', linestyle='-', linewidth=2)  
ax[0,1].set_xlim(0, 24)
ax[0,1].set_ylim([-1.25, 1.25])
ax[0,1].grid()
ax[0,1].set_ylabel('Normalized amplitude \n of rotation', family='serif', color='black', size=10)
#ax[0,1].set_xlabel('Length of the Bridge', family='serif', color='black', size=10)

line3, = ax[1,0].plot(LOB, MSrot3_smooth, color='#d48500', linewidth=1, linestyle='-', label='Mode I')    
ax[1,0].axhline(y=0, color='k', linestyle='-', linewidth=2)   
ax[1,0].set_xlim(0, 24)
ax[1,0].set_ylim([-1.25, 1.25])
ax[1,0].grid()
ax[1,0].set_ylabel('Normalized amplitude \n of rotation', family='serif', color='black', size=10)
ax[1,0].set_xlabel('Length of the Bridge', family='serif', color='black', size=10)

line4, = ax[1,1].plot(LOB, MSrot4_smooth, color='#d48500', linewidth=1, linestyle='-', label='Mode I')    
ax[1,1].axhline(y=0, color='k', linestyle='-', linewidth=2)   
ax[1,1].set_xlim(0, 24)
ax[1,1].set_ylim([-1.25, 1.25])
ax[1,1].grid()
ax[1,1].set_ylabel('Normalized amplitude \n of rotation', family='serif', color='black', size=10)
ax[1,1].set_xlabel('Length of the Bridge', family='serif', color='black', size=10)

def update(frame):
    total_frames = 200  # Total number of frames in the animation
    mid_frame = total_frames // 2  # Midpoint where the flip reverses

    # Flipping to a positive direction for the first half
    if frame <= mid_frame:
        factor = np.cos(frame * np.pi / mid_frame)
    # Returning to the original curve for the second half
    else:
        factor = np.cos((total_frames - frame) * np.pi / mid_frame)
    
    flipped_MS1 = MSrot1_smooth * factor
    flipped_MS2 = MSrot2_smooth * factor
    flipped_MS3 = MSrot3_smooth * factor
    flipped_MS4 = MSrot4_smooth * factor
    
    line1.set_ydata(flipped_MS1)
    line2.set_ydata(flipped_MS2)
    line3.set_ydata(flipped_MS3)
    line4.set_ydata(flipped_MS4)
    
    return line1, line2, line3, line4

# Create the animation
ani = FuncAnimation(fig, update, frames=200, interval=50, blit=True, repeat=True)
fig.tight_layout()
# Add text labels
plt.subplots_adjust(hspace=0.2)
plt.text(0.01, 0.99, "(a)", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.25, 0.99, "Mode I", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.5, 0.99, "(b)", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.75, 0.99, "Mode II", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.01, 0.5, "(c)", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.25, 0.5, "Mode III", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.5, 0.5, "(d)", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)
plt.text(0.75, 0.5, "Mode IV", fontdict=font2, fontweight='bold', transform=plt.gcf().transFigure)

# Save the animation
ani.save('RMS_anim_trial.mp4', writer='ffmpeg', fps=20)
writer = PillowWriter(fps=20)
ani.save('RMS_anim_trial.gif', writer=writer)

HTML(ani.to_jshtml())
```
